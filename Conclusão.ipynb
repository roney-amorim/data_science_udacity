```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("titanic-data-6.csv")

df.info()
```

```python
df_s = df[df['Survived'] == 1]
df_ds = df[df['Survived'] == 0]
df_ds_c = df_ds[df_ds['Pclass'] == 1]
df_c = df_s[df_s['Pclass'] == 1]
df_ds_c['Sex'].value_counts()
df_ds_c[df_ds_c['Age'] < 30]
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.hist(df_s['Pclass'], alpha=0.5, label='Sobreviveu')
ax.hist(df_ds['Pclass'], alpha=0.5, label='Não Sobreviveu')
ax.set_title('Quantos passageiros sobreviveram por classe')
ax.set_xlabel('Classe')
ax.set_ylabel('Counts')
ax.legend(loc='best')
plt.show()
```
