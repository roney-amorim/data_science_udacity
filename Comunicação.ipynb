```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline

df = pd.read_csv("titanic-data-6.csv")

df.info()
```

```python
ind = df['Survived'].value_counts().index
df['Survived'].value_counts()[ind].plot(kind="pie",figsize=(8,8))
```

```python
df.groupby(['Pclass','Survived'], as_index=False)['Age'].mean()
```

```python
df_m = df[df['Pclass'] == 1]
df_m = df.query('Pclass == 1 and Survived == 1')

# df_a = df[df['Age'] == '<20']
# df_a = df.query('Age == "<20"')
df_m
```

```python
classes = ['red','black']
survived_mean = df.groupby('Pclass')['Survived'].mean()
survived_mean.plot(kind='bar', title='Quantos passageiros sobreviveram', color=classes, alpha=0.7)
plt.xlabel('Survived',fontsize=18)
plt.ylabel('Pclass',fontsize=18)
```
